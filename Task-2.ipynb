```python
import pypsa
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

%matplotlib inline
plt.rc("figure", figsize=(8, 8))
network = pypsa.examples.ac_dc_meshed(from_master=True)
```

```python
network.optimize() 
```

<h1  style="color:withe;font-size:20;font-weight:700">Frankfurt </h1>

```python
df = pd.DataFrame(data={"Frankfurt Wind":network.generators_t.p["Frankfurt Wind"] ,"Frankfurt Gas": network.generators_t.p["Frankfurt Gas"]  ,
     "Total":network.generators_t.p["Frankfurt Wind"] + network.generators_t.p["Frankfurt Gas"]}).div(1e3)

ax = df.plot.area(stacked=False ,ylabel="GW",style=["-","-","--"]  ,linewidth=3 , figsize=(8,8))
```

```python

df = pd.DataFrame(data={"Frankfurt":network.loads_t.p_set["Frankfurt"] ,
     "Total":network.generators_t.p["Frankfurt Wind"] + network.generators_t.p["Frankfurt Gas"]}).div(1e3)

ax = df.plot.area(stacked=False ,ylabel="GW",style=["-","--"]  ,linewidth=3 , figsize=(8,8))
```

```python
network.generators_t.p
```

```python
df = pd.DataFrame(data={"Total Wind":network.generators_t.p["Manchester Wind"] +network.generators_t.p["Norway Wind"]  + network.generators_t.p["Frankfurt Wind"] ,
    "Total Gas": network.generators_t.p["Manchester Gas"] + network.generators_t.p["Norway Gas"] + network.generators_t.p["Frankfurt Gas"]  })

df["Total"] = df["Total Wind"]  +  df["Total Gas"]  
df = df.div(1e3)

ax = df.plot.area(stacked=False ,ylabel="GW",style=["-","-","--"]  ,linewidth=3 , figsize=(8,8))
```

<h1  style="color:withe;font-size:20;font-weight:700">Total loads and total p</h1>

```python
df1   = pd.DataFrame( data={"Total loads":network.loads_t.p_set["Norway"] + network.loads_t.p_set["Frankfurt"] + network.loads_t.p_set["Manchester"],
    "Total p":df["Total"].div(1e-3) } 
    ).div(1e3)

ax = df1.plot.area(stacked=False ,ylabel="GW",style=["-","--"]  ,linewidth=3 , figsize=(8,8))
```
